```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#dataset import
dataset=pd.read_csv('pizza.csv')
dataset
```

```python
dataset.isnull().any()
```

```python
dataset.shape
```

```python
dataset.head(2)
```

```python
#dataset splitting:training and testing
from sklearn.model_selection import train_test_split
x=dataset[['Size']]
y=dataset['Price']
xtrain, xtest, ytrain,ytest=train_test_split(x,y,test_size=.20, random_state=1)
```

```python
xtest
```

```python
xtrain
```

```python
#Import Linear Regression Model
from sklearn.linear_model import LinearRegression
#Create object for LinearRegression()
reg = LinearRegression()
#Train the model with xtrain and ytrain data
reg.fit(xtrain,ytrain)
#Test the model with x test data
reg.predict(xtest)
```

```python
ytest
```

```python
#Plotting the dataset
plt.scatter(dataset['Size'],dataset['Price'],marker='o',
            color='green')
plt.xlabel('Size in square inch')
plt.ylabel('Price in TK')
plt.title('pizza pricing')
plt.plot(dataset.Size, reg.predict(dataset[['Size']]))
```
