```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Sample dataset
data = {
    'Year': [2010, 2011, 2012, 2013, 2014, 2015],
    'Sales': [200, 220, 250, 270, 300, 350],
    'Profit': [50, 60, 70, 80, 90, 100]
}

# Convert the data into a pandas DataFrame
df = pd.DataFrame(data)

# -------------------------- Line Plot --------------------------

# Plotting a line chart
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['Sales'], marker='o', color='b', linestyle='-', label='Sales')
plt.plot(df['Year'], df['Profit'], marker='o', color='r', linestyle='-', label='Profit')

# Adding title and labels
plt.title('Sales and Profit Over the Years')
plt.xlabel('Year')
plt.ylabel('Amount (in units)')

# Show legend
plt.legend()

# Display the plot
plt.show()

# -------------------------- Bar Plot --------------------------

# Plotting a bar chart
plt.figure(figsize=(10, 6))
plt.bar(df['Year'], df['Sales'], color='green', label='Sales')
plt.bar(df['Year'], df['Profit'], color='orange', label='Profit', alpha=0.5)

# Adding title and labels
plt.title('Sales and Profit Bar Chart')
plt.xlabel('Year')
plt.ylabel('Amount')

# Show legend
plt.legend()

# Display the plot
plt.show()

# -------------------------- Scatter Plot --------------------------

# Plotting a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Year'], df['Sales'], color='blue', label='Sales')
plt.scatter(df['Year'], df['Profit'], color='red', label='Profit')

# Adding title and labels
plt.title('Sales and Profit vs Year')
plt.xlabel('Year')
plt.ylabel('Amount')

# Show legend
plt.legend()

# Display the plot
plt.show()

# -------------------------- Pie Chart --------------------------

# Example pie chart data
categories = ['Sales', 'Profit']
values = [df['Sales'].sum(), df['Profit'].sum()]

# Plotting a pie chart
plt.figure(figsize=(8, 8))
plt.pie(values, labels=categories, autopct='%1.1f%%', startangle=90, colors=['blue', 'orange'])
plt.title('Sales and Profit Distribution')

# Display the pie chart
plt.show()

# -------------------------- Seaborn Line Plot --------------------------

# Using seaborn for line plot
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Sales', data=df, label='Sales', color='blue')
sns.lineplot(x='Year', y='Profit', data=df, label='Profit', color='red')

# Adding title and labels
plt.title('Sales and Profit Over the Years (Seaborn)')
plt.xlabel('Year')
plt.ylabel('Amount (in units)')

# Show legend
plt.legend()

# Display the plot
plt.show()
```
