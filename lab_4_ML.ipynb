```python
import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [2,4,6,8]
plt.title("Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(x,y,marker='o',color='blue')
print(plt.show())
```

```python
x = [1,2,3,4]
y = [2,4,6,8]
plt.bar(x,y)
plt.title("Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
print(plt.show())
```

```python
x = [1,2,3,4]
y = [2,4,6,8]
plt.figure(figsize=(6,6))
plt.scatter(x,y,color='green')
plt.title("Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
print(plt.show())
```
